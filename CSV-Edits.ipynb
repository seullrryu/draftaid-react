```python
import pandas as pd
```

```python
players = pd.read_csv('values-players.csv')
fantasy_pros = pd.read_csv('FantasyPros_2023_Draft_ALL_Rankings.csv')
```

```python
fantasy_pros.head()
```

```python
fantasy_pros = fantasy_pros.rename(columns={"PLAYER NAME":"player", "BYE WEEK": "bye"})
fantasy_pros.head()
```

```python
fantasy_pros = fantasy_pros[['RK','TIERS','player','TEAM','POS','bye']]
fantasy_pros.head()
```

```python
players.head()
```

```python
players = players[['player', 'age', 'draft_year']]
players.head()
```

```python
cleaned_players = pd.merge(players, fantasy_pros, on="player", how="inner")
cleaned_players.head()
```

```python
json = cleaned_players.to_json(orient='records')
print(json)
```
